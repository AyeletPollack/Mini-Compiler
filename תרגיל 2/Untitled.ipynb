```python
%%cython

import os
import os.path
                
# shir bracya 207447319
# Ayelet Pollack 206935330
# 150060.21.5780.42

#global 
counter = "A"

#PUSH
cdef push(fileAsm, line, fileName):
    #PUSH-CONSTANT
    if line.startswith("constant", 5, 13):
        num = line.split()[2]
        fileAsm.write("@"+num+"\n")
        fileAsm.write("D=A\n")
        fileAsm.write("@SP\n")
        fileAsm.write("A=M\n")
        fileAsm.write("M=D\n")
        fileAsm.write("@SP\n")
        fileAsm.write("M=M+1\n")
     #PUSH-LOCAL
    if line.startswith("local", 5, 10):
        num = line.split()[2]
        fileAsm.write("@"+num+"\n")
        fileAsm.write("D=A\n")
        fileAsm.write("@LCL\n")
        fileAsm.write("A=M+D\n")
        fileAsm.write("D=M\n")
        fileAsm.write("@SP\n")
        fileAsm.write("A=M\n")
        fileAsm.write("M=D\n")
        fileAsm.write("@SP\n")
        fileAsm.write("M=M+1\n")
    #PUSH-ARGUMENT
    if line.startswith("argument", 5, 13):
        num = line.split()[2]
        fileAsm.write("@"+num+"\n")
        fileAsm.write("D=A\n")
        fileAsm.write("@ARG\n")
        fileAsm.write("A=M+D\n")
        fileAsm.write("D=M\n")
        fileAsm.write("@SP\n")
        fileAsm.write("A=M\n")
        fileAsm.write("M=D\n")
        fileAsm.write("@SP\n")
        fileAsm.write("M=M+1\n")
    #PUSH-THAT
    if line.startswith("that", 5, 9):
        num = line.split()[2]
        fileAsm.write("@"+num+"\n")
        fileAsm.write("D=A\n")
        fileAsm.write("@THAT\n")
        fileAsm.write("A=M+D\n")
        fileAsm.write("D=M\n")
        fileAsm.write("@SP\n")
        fileAsm.write("A=M\n")
        fileAsm.write("M=D\n")
        fileAsm.write("@SP\n")
        fileAsm.write("M=M+1\n") 
    #PUSH-THIS
    if line.startswith("this", 5, 9):
        num = line.split()[2]
        fileAsm.write("@"+num+"\n")
        fileAsm.write("D=A\n")
        fileAsm.write("@THIS\n")
        fileAsm.write("A=M+D\n")
        fileAsm.write("D=M\n")
        fileAsm.write("@SP\n")
        fileAsm.write("A=M\n")
        fileAsm.write("M=D\n")
        fileAsm.write("@SP\n")
        fileAsm.write("M=M+1\n")
    #PUSH-TEMP
    if line.startswith("temp", 5, 9):
        num = int(line.split()[2])
        if((num>=0) and (num<=7)):
            if(num==0):temp="5"
            if(num==1):temp="6"
            if(num==2):temp="7"
            if(num==3):temp="8"
            if(num==4):temp="9"
            if(num==5):temp="10"
            if(num==6):temp="11"
            if(num==7):temp="12"
            fileAsm.write("@"+temp+"\n")
            fileAsm.write("D=A\n")
            fileAsm.write("@TEMP\n")
            fileAsm.write("A=M+D\n")
            fileAsm.write("D=M\n")
            fileAsm.write("@SP\n")
            fileAsm.write("A=M\n")
            fileAsm.write("M=D\n")
            fileAsm.write("@SP\n")
            fileAsm.write("M=M+1\n")
        else: print("ERROR")
    #PUSH-POINTER
    if line.startswith("pointer", 5, 12):
        num = int(line.split()[2])
        #PUSH-POINTER 0
        if (num==0):
            fileAsm.write("@THIS\n")
            fileAsm.write("D=M\n")
            fileAsm.write("@SP\n")
            fileAsm.write("A=M\n")
            fileAsm.write("M=D\n")
            fileAsm.write("@SP\n")
            fileAsm.write("M=M+1\n")
        #PUSH-POINTER 1
        elif (num==1):
            fileAsm.write("@THAT\n")
            fileAsm.write("D=M\n")
            fileAsm.write("@SP\n")
            fileAsm.write("A=M\n")
            fileAsm.write("M=D\n")
            fileAsm.write("@SP\n")
            fileAsm.write("M=M+1\n")
    #PUSH-STATIC
    if line.startswith("static", 5, 11):
        num = line.split()[2]
        fileAsm.write("@"+fileName+"."+num+"\n")
        fileAsm.write("D=M\n")
        fileAsm.write("@SP\n")
        fileAsm.write("A=M\n")
        fileAsm.write("M=D\n")
        fileAsm.write("@SP\n")
        fileAsm.write("M=M+1\n")
    
#POP
cdef pop(fileAsm, line, fileName):
    #POP-LOCAL
    if line.startswith("local", 4, 9):
        num = int(line.split()[2])
        fileAsm.write("@SP\n")
        fileAsm.write("A=M-1\n")
        fileAsm.write("D=M\n")
        fileAsm.write("@LCL\n")
        fileAsm.write("A=M\n")
        while (num>0):
            fileAsm.write("A=A+1\n")
            num=num-1
        fileAsm.write("M=D\n")
        fileAsm.write("@SP\n")
        fileAsm.write("M=M-1\n")
    #POP-THIS
    if line.startswith("this", 4, 8):
        num = int(line.split()[2])
        fileAsm.write("@SP\n")
        fileAsm.write("A=M-1\n")
        fileAsm.write("D=M\n")
        fileAsm.write("@THIS\n")
        fileAsm.write("A=M\n")
        while (num>0):
            fileAsm.write("A=A+1\n")
            num=num-1
        fileAsm.write("M=D\n")
        fileAsm.write("@SP\n")
        fileAsm.write("M=M-1\n")
    #POP-ARGUMENT
    if line.startswith("argument", 4, 12):
        num = int(line.split()[2])
        fileAsm.write("@SP\n")
        fileAsm.write("A=M-1\n")
        fileAsm.write("D=M\n")
        fileAsm.write("@ARG\n")
        fileAsm.write("A=M\n")
        while (num>0):
            fileAsm.write("A=A+1\n")
            num=num-1
        fileAsm.write("M=D\n")
        fileAsm.write("@SP\n")
        fileAsm.write("M=M-1\n")
    #POP-THAT
    if line.startswith("that", 4, 8):
        num = int(line.split()[2])
        fileAsm.write("@SP\n")
        fileAsm.write("A=M-1\n")
        fileAsm.write("D=M\n")
        fileAsm.write("@THAT\n")
        fileAsm.write("A=M\n")
        while (num>0):
            fileAsm.write("A=A+1\n")
            num=num-1
        fileAsm.write("M=D\n")
        fileAsm.write("@SP\n")
        fileAsm.write("M=M-1\n")
    #POP-TEMP
    if line.startswith("temp", 4, 8):
        num = int(line.split()[2])
        if((num>=0) and (num<=7)):
            fileAsm.write("@SP\n")
            fileAsm.write("A=M-1\n")
            fileAsm.write("D=M\n")
            if(num==0):temp="5"
            if(num==1):temp="6"
            if(num==2):temp="7"
            if(num==3):temp="8"
            if(num==4):temp="9"
            if(num==5):temp="10"
            if(num==6):temp="11"
            if(num==7):temp="12"
            fileAsm.write("@"+temp+"\n")
            fileAsm.write("M=D\n")
            fileAsm.write("@SP\n")
            fileAsm.write("M=M-1\n")
        else: print("ERROR")
    #POP-POINTER
    if line.startswith("pointer", 4, 11):
        num = int(line.split()[2])
        #POP-POINTER 0
        if (num==0):
            fileAsm.write("@SP\n")
            fileAsm.write("A=M-1\n")
            fileAsm.write("D=M\n")
            fileAsm.write("@THIS\n")
            fileAsm.write("M=D\n")
            fileAsm.write("@SP\n")
            fileAsm.write("M=M-1\n")
        #POP-POINTER 1
        elif (num==1):
            fileAsm.write("@SP\n")
            fileAsm.write("A=M-1\n")
            fileAsm.write("D=M\n")
            fileAsm.write("@THAT\n")
            fileAsm.write("M=D\n")
            fileAsm.write("@SP\n")
            fileAsm.write("M=M-1\n")
    #POP-STATIC
    if line.startswith("static", 4, 10):
        num = line.split()[2]
        fileAsm.write("@SP\n")
        fileAsm.write("A=M-1\n")
        fileAsm.write("D=M\n")
        fileAsm.write("@"+fileName+"."+num+"\n")
        fileAsm.write("M=D\n")
        fileAsm.write("@SP\n")
        fileAsm.write("M=M-1\n")
        
#ADD    
cdef add(fileAsm):
    fileAsm.write("@SP\n")
    fileAsm.write("A=M-1\n")
    fileAsm.write("D=M\n")
    fileAsm.write("A=A-1\n")
    fileAsm.write("M=D+M\n")
    fileAsm.write("@SP\n")
    fileAsm.write("M=M-1\n")
    
#SUB  
cdef sub(fileAsm):
    fileAsm.write("@SP\n")
    fileAsm.write("A=M-1\n")
    fileAsm.write("D=M\n")
    fileAsm.write("A=A-1\n")
    fileAsm.write("M=M-D\n")
    fileAsm.write("@SP\n")
    fileAsm.write("M=M-1\n")
    
#EQ  
cdef eq(fileAsm):
    fileAsm.write("@SP\n")
    fileAsm.write("M=M-1\n")
    fileAsm.write("@SP\n")
    fileAsm.write("A=M\n")
    fileAsm.write("D=M\n")
    fileAsm.write("@SP\n")
    fileAsm.write("M=M-1\n")
    fileAsm.write("@SP\n")
    fileAsm.write("A=M\n")
    fileAsm.write("A=M\n")
    fileAsm.write("D=D-A\n")
    fileAsm.write("@IF_TRUE0\n")
    fileAsm.write("D;JEQ\n")
    fileAsm.write("@IF_FALSE0\n")
    fileAsm.write("D=0\n")
    fileAsm.write("D;JEQ\n")
    fileAsm.write("(IF_TRUE0)\n")
    fileAsm.write("D=-1\n")
    fileAsm.write("(IF_FALSE0)\n")
    fileAsm.write("@SP\n")
    fileAsm.write("A=M\n")
    fileAsm.write("M=D\n")
    fileAsm.write("@SP\n")
    fileAsm.write("M=M+1\n")
    
#LT 
cdef lt(fileAsm):
    fileAsm.write("@SP\n")
    fileAsm.write("M=M-1\n")
    fileAsm.write("@SP\n")
    fileAsm.write("A=M\n")
    fileAsm.write("D=M\n")
    fileAsm.write("@SP\n")
    fileAsm.write("M=M-1\n")
    fileAsm.write("@SP\n")
    fileAsm.write("A=M\n")
    fileAsm.write("A=M\n")
    fileAsm.write("D=D-A\n")
    fileAsm.write("@IF_TRUE1\n")
    fileAsm.write("D;JGT\n")
    fileAsm.write("@IF_FALSE1\n")
    fileAsm.write("D=0\n")
    fileAsm.write("D;JEQ\n")
    fileAsm.write("(IF_TRUE1)\n")
    fileAsm.write("D=-1\n")
    fileAsm.write("(IF_FALSE1)\n")
    fileAsm.write("@SP\n")
    fileAsm.write("A=M\n")
    fileAsm.write("M=D\n")
    fileAsm.write("@SP\n")
    fileAsm.write("M=M+1\n")
    
#GT 
cdef gt(fileAsm):
    fileAsm.write("@SP\n")
    fileAsm.write("M=M-1\n")
    fileAsm.write("@SP\n")
    fileAsm.write("A=M\n")
    fileAsm.write("D=M\n")
    fileAsm.write("@SP\n")
    fileAsm.write("M=M-1\n")
    fileAsm.write("@SP\n")
    fileAsm.write("A=M\n")
    fileAsm.write("A=M\n")
    fileAsm.write("D=A-D\n")
    fileAsm.write("@IF_TRUE2\n")
    fileAsm.write("D;JGT\n")
    fileAsm.write("@IF_FALSE2\n")
    fileAsm.write("D=0\n")
    fileAsm.write("D;JEQ\n")
    fileAsm.write("(IF_TRUE2)\n")
    fileAsm.write("D=-1\n")
    fileAsm.write("(IF_FALSE2)\n")
    fileAsm.write("@SP\n")
    fileAsm.write("A=M\n")
    fileAsm.write("M=D\n")
    fileAsm.write("@SP\n")
    fileAsm.write("M=M+1\n")
    
#AND
cdef andfunc(fileAsm): 
    fileAsm.write("@SP\n")
    fileAsm.write("M=M-1\n")
    fileAsm.write("@SP\n")
    fileAsm.write("A=M\n")
    fileAsm.write("D=M\n")
    fileAsm.write("@SP\n")
    fileAsm.write("M=M-1\n")
    fileAsm.write("@SP\n")
    fileAsm.write("A=M\n")
    fileAsm.write("A=M\n")
    fileAsm.write("D=D&A\n")
    fileAsm.write("@SP\n")
    fileAsm.write("A=M\n")
    fileAsm.write("M=D\n")
    fileAsm.write("@SP\n")
    fileAsm.write("M=M+1\n")
    
#OR
cdef orfunc(fileAsm):
    fileAsm.write("@SP\n")
    fileAsm.write("M=M-1\n")
    fileAsm.write("@SP\n")
    fileAsm.write("A=M\n")
    fileAsm.write("D=M\n")
    fileAsm.write("@SP\n")
    fileAsm.write("M=M-1\n")
    fileAsm.write("@SP\n")
    fileAsm.write("A=M\n")
    fileAsm.write("A=M\n")
    fileAsm.write("D=D|A\n")
    fileAsm.write("@SP\n")
    fileAsm.write("A=M\n")
    fileAsm.write("M=D\n")
    fileAsm.write("@SP\n")
    fileAsm.write("M=M+1\n")
    
#NEG
cdef neg(fileAsm):
    fileAsm.write("@SP\n")
    fileAsm.write("A=M-1\n")
    fileAsm.write("D=M\n")
    fileAsm.write("M=-D\n")
    
#NOT
cdef _not(fileAsm):
    fileAsm.write("@SP\n")
    fileAsm.write("A=M-1\n")
    fileAsm.write("D=M\n")
    fileAsm.write("M=!D\n")
    
#LABEL
cdef label(fileAsm, line, fileName):
        c = line.split()[1]
        fileAsm.write("("+fileName+"."+c+")"+"\n")

#IF-GOTO
cdef if_goto(fileAsm, line, fileName):
        c = line.split()[1]
        fileAsm.write("@SP\n")
        fileAsm.write("M=M-1\n")
        fileAsm.write("A=M\n")
        fileAsm.write("D=M\n")
        fileAsm.write("@"+fileName+"."+c+"\n")
        fileAsm.write("D;JNE\n")
        
#GOTO
cdef goto(fileAsm, line, fileName):
        c = line.split()[1]
        fileAsm.write("@"+fileName+"."+c+"\n")
        fileAsm.write("0;JMP\n")
        
#RETURN
cdef _return(fileAsm):
    #FRAME = LCL
    fileAsm.write("@LCL\n")
    fileAsm.write("D=M\n")
    #RET = *(FRAME-5)
    #RAM[13] = (LOCAL-5)
    fileAsm.write("@5\n")
    fileAsm.write("A=D-A\n")
    fileAsm.write("D=M\n")
    fileAsm.write("@13\n")
    fileAsm.write("M=D\n")
    #*ARG = POP()
    fileAsm.write("@SP\n")
    fileAsm.write("M=M-1\n")
    fileAsm.write("A=M\n")
    fileAsm.write("D=M\n")
    fileAsm.write("@ARG\n")
    fileAsm.write("A=M\n")
    fileAsm.write("M=D\n")
    #SP = ARG+1
    fileAsm.write("@ARG\n")
    fileAsm.write("D=M\n")
    fileAsm.write("@SP\n")
    fileAsm.write("M=D+1\n")
    #THAT = *(FRAM-1)
    fileAsm.write("@LCL\n")
    fileAsm.write("M=M-1\n")
    fileAsm.write("A=M\n")
    fileAsm.write("D=M\n")
    fileAsm.write("@THAT\n")
    fileAsm.write("M=D\n")
    #THIS = *(FRAM-2)
    fileAsm.write("@LCL\n")
    fileAsm.write("M=M-1\n")
    fileAsm.write("A=M\n")
    fileAsm.write("D=M\n")
    fileAsm.write("@THIS\n")
    fileAsm.write("M=D\n")
    #ARG = *(FRAM-3)
    fileAsm.write("@LCL\n")
    fileAsm.write("M=M-1\n")
    fileAsm.write("A=M\n")
    fileAsm.write("D=M\n")
    fileAsm.write("@ARG\n")
    fileAsm.write("M=D\n")
    #LCL = *(FRAM-4)
    fileAsm.write("@LCL\n")
    fileAsm.write("M=M-1\n")
    fileAsm.write("A=M\n")
    fileAsm.write("D=M\n")
    fileAsm.write("@LCL\n")
    fileAsm.write("M=D\n")
    #GOTO RET
    fileAsm.write("@13\n")
    fileAsm.write("A=M\n")
    fileAsm.write("0;JMP\n")

#CALL
cdef call(fileAsm, line):
    global counter
    g = line.split()[1]
    n = line.split()[2]
    #push return-address
    fileAsm.write("@"+g+n+".ReturnAddress"+counter+"\n")
    fileAsm.write("D=A\n")
    fileAsm.write("@SP\n")
    fileAsm.write("A=M\n")
    fileAsm.write("M=D\n")
    fileAsm.write("@SP\n")
    fileAsm.write("M=M+1\n")
    #push LCL
    fileAsm.write("@LCL\n")
    fileAsm.write("D=M\n")
    fileAsm.write("@SP\n")
    fileAsm.write("A=M\n")
    fileAsm.write("M=D\n")
    fileAsm.write("@SP\n")
    fileAsm.write("M=M+1\n")
    #push ARG
    fileAsm.write("@ARG\n")
    fileAsm.write("D=M\n")
    fileAsm.write("@SP\n")
    fileAsm.write("A=M\n")
    fileAsm.write("M=D\n")
    fileAsm.write("@SP\n")
    fileAsm.write("M=M+1\n")
    #push THIS
    fileAsm.write("@THIS\n")
    fileAsm.write("D=M\n")
    fileAsm.write("@SP\n")
    fileAsm.write("A=M\n")
    fileAsm.write("M=D\n")
    fileAsm.write("@SP\n")
    fileAsm.write("M=M+1\n")
    #PUSH THAT
    fileAsm.write("@THAT\n")
    fileAsm.write("D=M\n")
    fileAsm.write("@SP\n")
    fileAsm.write("A=M\n")
    fileAsm.write("M=D\n")
    fileAsm.write("@SP\n")
    fileAsm.write("M=M+1\n")
    #ARG = SP-n-5
    fileAsm.write("@SP\n")
    fileAsm.write("D=M\n")
    temp = int(n)+5
    fileAsm.write("@"+str(temp)+"\n")
    fileAsm.write("D=D-A\n")
    fileAsm.write("@ARG\n")
    fileAsm.write("M=D\n")
    #LCL = SP
    fileAsm.write("@SP\n")
    fileAsm.write("D=M\n")
    fileAsm.write("@LCL\n")
    fileAsm.write("M=D\n")
    #GOTO G
    fileAsm.write("@"+g+"\n")
    fileAsm.write("0;JMP\n")
    #label return-address
    fileAsm.write("("+g+n+".ReturnAddress"+counter+")"+"\n")
    counter = chr(ord(counter) + 1)
    
#INIT
cdef init(fileAsm):
    fileAsm.write("@256\n")
    fileAsm.write("D=A\n")
    fileAsm.write("@SP\n")
    fileAsm.write("M=D\n")
    sys = "call Sys.init 0"
    call(fileAsm,sys)
    
#FUNCTION
cdef function(fileAsm, line):
    g = line.split()[1]
    k = line.split()[2]
    #label
    fileAsm.write("("+g+")\n")
    #K
    fileAsm.write("@"+k+"\n")
    fileAsm.write("D=A\n")
    fileAsm.write("@"+g+k+".END\n")
    fileAsm.write("D;JEQ\n")
    fileAsm.write("("+g+k+".LOOP)\n")
    fileAsm.write("@SP\n")
    fileAsm.write("A=M\n")
    fileAsm.write("M=0\n")
    fileAsm.write("@SP\n")
    fileAsm.write("M=M+1\n")
    fileAsm.write("@"+g+k+".LOOP\n")
    fileAsm.write("D=D-1;JNE\n")
    fileAsm.write("("+g+k+".END)\n")

    

#MAIN ------------------------------------------------------------------- MAIN
#MAIN ------------------------------------------------------------------- MAIN            
#MAIN ------------------------------------------------------------------- MAIN
path = input()
print("\n")
strt = (path).split("\\")[-1]
temp = path+"\\"
file1 = open(temp+strt+".asm", "w")
init(file1)
for filenames in os.listdir(path):
    if(".vm" in filenames):
        with open(os.path.join(path, filenames)) as file:
            if(".vm" in file.name):
                strt = (file.name).split("\\")[-1]
                for line in file:
                    if line.startswith("push"):
                        push(file1, line, strt)
                    if line.startswith("add"):
                        add(file1)
                    if line.startswith("pop"):
                        pop(file1, line, strt)
                    if line.startswith("sub"):
                        sub(file1)
                    if line.startswith("and"):
                        andfunc(file1)
                    if line.startswith("or"):
                        orfunc(file1)
                    if line.startswith("eq"):
                        eq(file1)
                    if line.startswith("lt"):
                        lt(file1)
                    if line.startswith("gt"):
                        gt(file1)
                    if line.startswith("neg"):
                        neg(file1)
                    if line.startswith("not"):
                        _not(file1)
                    if line.startswith("label"):
                        label(file1, line, strt)
                    if line.startswith("if-goto"):
                        if_goto(file1, line, strt)
                    if line.startswith("goto"):
                        goto(file1, line, strt)
                    if line.startswith("return"):
                        _return(file1)
                    if line.startswith("call"):
                        call(file1, line)
                    if line.startswith("function"):
                        function(file1, line)
                    
file1.close()

    

                
```
